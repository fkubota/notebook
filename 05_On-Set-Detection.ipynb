# Import

```python
import numpy as np
import librosa
from scipy.io import wavfile
import matplotlib.pyplot as plt
```

```python
wav, sr = librosa.load('./05_On-Set-Detection/loop1.wav')
```

```python
onset_frames = librosa.onset.onset_detect(y=wav, sr=sr)
onset = librosa.frames_to_time(onset_frames, sr=sr)
```

```python
plt.figure(figsize=(30, 5))
plt.plot(np.arange(0, len(wav))/sr, wav, alpha=0.5)
plt.vlines(onset, -1, 1, color='r', alpha=0.9, linestyle='-', label='Onsets')
```
